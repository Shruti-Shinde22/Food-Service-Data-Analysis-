```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv("data_set_.csv")
```

```python
dataframe
```

```python
# data cleaning

def handleRate(Value):
    Value = str(Value).split('/') 
    Value = Value[0];
    return float(Value)

dataframe['rate'] = dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("type of restaurant")
```

```python
# Conclusion: max people visit Dinning restaurent
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes':grouped_data})
plt.plot(result,c="green",marker="o")
plt.xlabel("Type of restaurant", c="red",size=20)
plt.ylabel("votes",c="red",size=20)
```

```python
# conclusion: max votes for dinning restaurant
```

```python
plt.hist(dataframe['rate'],bins=5)
plt.title("Ratings /distribution")
plt.show()
```

```python
# conclusion: highest rating 3.75-4.25
```

```python
couple_data = dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(x='online_order', y = 'rate', data=dataframe)
```

```python
# conclusion:- offline order received lower ratings
```

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)',columns='online_order',aggfunc = 'size',fill_value=0)
sns.heatmap(pivot_table,annot=True,cmap="YlGnBu",fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (type)")
plt.show()
```

```python
# conclusion: Dinning restaurants primarily offers offline order
#             Cafes primarily receive online orders
#             client prefers in person at restaurant, but prefer online ordering at cafes
```
